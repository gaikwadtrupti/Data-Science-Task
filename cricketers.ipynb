# How many cricketers are still alive?


```python
from datetime import datetime
```

```python

df['Birthdate'] = pd.to_datetime(df['Birthdate'])
df['Date_of_death'] = pd.to_datetime(df['Date_of_death']) 
current_date = datetime.now()


alive_cricketers = df[(df['Date_of_death'].isnull()) | (df['Date_of_death'] > current_date)]


num_alive_cricketers = alive_cricketers.shape[0]


print("Number of cricketers still alive:", num_alive_cricketers)
```

# Plot the counts of players from each country

```python

country_counts = df['COUNTRY'].value_counts()


```

```python

plt.figure(figsize=(12, 7))
country_counts.plot(kind='bar')
plt.title('Counts of Players from Each Country')
plt.xlabel('COUNTRY')
plt.ylabel('Number of Players')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Plot the data for average ODI batting average of players from India, Pakistan, Sri Lanka and Bangladesh

```python

filtered_df = df[df['COUNTRY'].isin(['India', 'Pakistan', 'Sri Lanka', 'Bangladesh'])]



average_batting_average = filtered_df.groupby('COUNTRY')['BATTING_ODIs_Ave'].mean()

# Plotting the data
plt.figure(figsize=(10, 6))
average_batting_average.plot(kind='bar', color=['blue', 'orange', 'red', 'green'])
plt.title('Average ODI Batting Average of Players')
plt.xlabel('Country')
plt.ylabel('Average Batting Average')
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

# Find all the players from Indian and Pakistan whose age is between 60 to 70.
